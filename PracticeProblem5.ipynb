```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def extract_prices(filename, year):
    raw_data = list()
    with open(filename, mode='r', newline='') as data:
        reader = csv.reader(data)
        for r in reader:
            if r[0].startswith(year): 
                for item in r[1:5]:
                    raw_data.append(float(item))
    
    data = np.array(raw_data).reshape(int(len(raw_data)/4),4)
    
    o = data[:,:1]
    h = data[:,1:2]
    l = data[:,2:3]
    c = data[:,3:4]
    
    return o, h, l, c
                
```

```python
openP, highP, lowP, closeP =  extract_prices('goog.us.txt','2016')
```
